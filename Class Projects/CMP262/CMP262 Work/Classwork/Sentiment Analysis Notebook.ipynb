# Grapical Display of Songs Lyrics with a Word Cloud

```python
!pip install wordcloud
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
wordcloud = WordCloud().generate(document)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")
wordcloud = WordCloud(background_color = "white", max_words = 2, stopwords = stopwords).generate(document)
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
```
