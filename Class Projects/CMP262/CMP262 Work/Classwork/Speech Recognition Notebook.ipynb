```python
import speech_recognition as sr
import pyttsx3
```

```python
r=sr.Recognizer()
```

## Speech to Text, then Text to Speech
### Input Source: Microphone
#### Create a method that can speak, given text

```python
def SpeakText(command):
    engine = pyttsx3.init()
    engine.say(command)
    engine.runAndWait()
```

### The following code connexts to the microphone, listens for 3 seconds, then converts the audio to text. It then repeats back the text that it heard

```python
with sr.Microphone() as source2:
    r.adjust_for_ambient_noise(source2, duration = 1)
    audio2 = r.listen(source2, timeout = 3)
    MyText = r.recognize_google(audio2)
    MyText = MyText.lower()
    
    print("Did you say " + MyText)
    SpeakText(MyText)
```

```python
filename = "test.wav"
```

```python
with sr.AudioFile(filename) as source:
    audio_data = r.record(source)
    text = r.recognize_google(audio_data)
    print(text)
```
